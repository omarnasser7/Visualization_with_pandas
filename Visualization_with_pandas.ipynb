## Data Visualization with Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_train.csv')
df.head()
```

### plot()
* It make chars - defaults is line plot
* More graphs with kind='line', 'bar', 'scatter', 'hist', 'etc..'

#### line plot

```python
df.plot()
```

#### scatter plot

```python
df.plot.scatter(x='exam', y='grade')
plt.savefig('my_plot1.png')
```

#### Histogram

```python
df.plot.hist()
```

#### Area plot

```python
df.plot.area()
```

#### Bar Char

```python
df.plot.bar()
```

```python
df.plot.bar(stacked=True)
```

```python
df.plot.barh()
```

#### Box Plot

```python
df.plot.box()
```

#### Hexagonal binning plot

```python
df.plot.hexbin(x='exam', y='grade', gridsize=20)
```

#### Density Estimate plot 

```python
df.plot.kde()
```

#### pie plot

```python
df.plot.pie(y='grade', autopct='%1.1f%%')
```

```python
plt.savefig('my_plot.png')
```
